```python
__author__ = 'IGOR'
import pandas as pd
from datetime import datetime
import numpy as np

raw_data = pd.read_csv('trajectory_data_sample.csv')
```

```python
def get_timestamp(row):
    return datetime.strptime(row, '%d.%m.%Y %H:%M')

raw_data['timestamp'] = raw_data['date_event'].apply( get_timestamp ) 
raw_data['delta'] = (raw_data['timestamp'] - raw_data['timestamp'].shift(1).fillna(0).astype('datetime64[ns]'))
```

```python
raw_data.head().delta < 60*10
```

```python
raw_data.head(20)
```

```python
raw_data.delta = raw_data.delta / 1000000000# converting to seconds
```

```python
raw_data['is_session'] = ( raw_data.delta < 60*10 ) # consider session, if there is less then 10 minutes of break between events
```

```python
raw_data.is_session.head(10)
```

```python
row_iterator = raw_data.iterrows()
 # take first item from row_iterator

    
last = row_iterator.next()
#print "lst", last
'''
for row in row_iterator:
    print "row", row[0]
    #last = row
    print "last", last[0]
    last = row
'''

session_list = [0]
session_id = 0
is_click = []
for row in row_iterator:
    next = row[1][['is_session','event']]
     
    #print "current:", curr
    #print last[1]
    
    curr = last[1][['is_session','event']]
    #print next['event']
    #print "next:", next
    if curr['is_session']:
        if next['is_session']:
            pass
        else:
            session_id +=1
    else:
        if next['is_session']:
            pass
        else:
            session_id +=1
    session_list.append(session_id)        
    
    #print "events " + curr['event'] + ' ' + next['event']
    
    if curr['event'] == 'visit':
        #print next['event']
        if next['event'] == 'click':
            is_click.append(1)
            #print curr,'visit+click'
        else:
            is_click.append(0)
    else:
        is_click.append(0)
    last = row        
    

is_click.append(0)
```

```python
len(is_click)
#len(session_list)
raw_data['session_id'] = session_list
raw_data['is_click'] = is_click
```

```python
processed_data = raw_data[raw_data.event != "click"] # transforming data to have a click as a target variable in a separate column
```

```python
processed_data = pd.concat([processed_data, pd.get_dummies(processed_data['channel'])],axis=1) # transform category column into binary variables
```

```python
#print raw_data[['date_event','event','session_id', 'is_click']]
''' add columns:
aggregate by user:
amount of visits by channel type
amount of clicks by channel type

aggregate by session:
amount of visits by channel type
amount of clicks by channel type
['is_click', 'Affiliation', 'Brand', 'Direct Access', 'Display', 'E-mail', 'Loyalty email', 'Other', 'Partners', 'Referrer', 'SEM', 'SEO', 'Social networks']
'''
```

```python
# columns (sum of visits by each column):
# Direct Access, Other etc.
# sum of "is_click"
# add columns to main DF with this information
# by cust_id, by is_session
channels = ['is_click', 'Affiliation', 'Brand', 'Direct Access', 'Display', 'E-mail', 'Loyalty email', 'Other', 'Partners', 'Referrer', 'SEM', 'SEO', 'Social networks']

session_data = processed_data.groupby('session_id').sum()[channels]
user_data = processed_data.groupby('cust_id').sum()[channels]
```

```python
processed_data.head()
```

```python
#['is_click', 'Affiliation', 'Brand', 'Direct Access', 'Display', 'E-mail', 'Loyalty email', 'Other', 'Partners', 'Referrer', 'SEM', 'SEO', 'Social networks']

session_data.columns = ["Session-"+x for x in channels]
user_data.columns = ["User-"+x for x in channels]
```

```python
session_data[session_data['Session-is_click'] > 10]
```

```python
processed_data = processed_data.merge(user_data, left_on='cust_id', right_index=True)
```

```python
processed_data = processed_data.merge(session_data, left_on='session_id', right_index=True)
```

```python
processed_data.head(200).to_csv('processed_data.csv')
```

```python
features = ['is_click', 'Affiliation', 'Brand', 'Direct Access', 'Display', 'E-mail', 'Loyalty email', 'Other', 'Partners', 'Referrer', 'SEM', 'SEO', 'Social networks']

session_data.columns = ["Session-"+x for x in channels]
user_data.columns = ["User-"+x for x in channels]

features.extend(session_data)
features.extend(user_data)

train_data = processed_data[features] #'cust_id','session_id',
```

```python
data = np.array(train_data)
data.shape
```

```python
Y = data[:,[0]]
X = data[:,1:]
```

```python
from sklearn.cross_validation import train_test_split,cross_val_score
from sklearn.linear_model import LogisticRegression
train_X, test_X, train_Y, test_Y = train_test_split(X,Y, test_size=0.33, random_state=42)
```

```python
est = LogisticRegression()
scores = cross_val_score(est, train_X, train_Y, cv=5)
#cross_val_score
```

```python
scores
```

```python
est.fit(train_X, train_Y)
predicted_Y = est.predict(test_X)
predicted_prob_Y = est.predict_proba(test_X)
```

```python
predicted_Y.shape
```

```python
pred_Y = predicted_Y.reshape(52738, 1)
pred_prob_Y = predicted_prob_Y[:,1].reshape(52738, 1)
```

```python
from sklearn.metrics import mean_squared_error, precision_recall_fscore_support, precision_recall_curve, f1_score, roc_auc_score, precision_recall_curve, auc, roc_curve
```

```python
mean_squared_error(test_Y, pred_Y)
```

```python
precision_recall_fscore_support(test_Y, pred_Y)
```

```python
f1_score(test_Y, pred_Y)
```

```python
roc_auc_score(test_Y, pred_Y)
```

```python
import pylab as pl
```

```python
# Compute Precision-Recall and plot curve
precision, recall, thresholds = precision_recall_curve(test_Y, pred_prob_Y)
area = auc(recall, precision)
print("Area Under Curve: %0.2f" % area)

pl.clf()
pl.plot(recall, precision, label='Precision-Recall curve')
pl.xlabel('Recall')
pl.ylabel('Precision')
pl.ylim([0.0, 1.0])
pl.xlim([0.0, 1.0])
pl.title('Precision-Recall example: AUC=%0.2f' % area)
pl.legend(loc="lower left")
pl.show()
```
